```python
def is_subset_tags(extracted_tags_str: str, original_tags_str: str):
    """
    Check whether all extracted category tags are present in the original tags.
    Both inputs are pipe-separated strings.
    Returns None if extracted tags are empty.
    """
    # ext  = {t.strip().lower() for t in extracted_tags_str.split("|") if t.strip()}
    ext = {t.strip().lower() for t in extracted_tags_str.replace("[","").replace("]","").split(",") if t.strip()}
    orig = {t.strip().lower() for t in original_tags_str.replace("[","").replace("]","").replace("'","").split(",") if t.strip()}

    print(ext)
    print(orig)

    if not ext:
        return None
    return ext.issubset(orig)

print(is_subset_tags("Choir", "['Choir', 'Music instructor']"))  # True
```
